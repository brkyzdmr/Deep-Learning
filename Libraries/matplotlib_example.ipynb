## Data Visualization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
# or plt.show()
```

```python
x = np.arange(0, 10)
y = x**2
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, 'r--')
```

```python
plt.plot(x, y)
plt.xlim(0,4)
plt.ylim(0,4)
plt.title("title")
plt.xlabel("x-label")
plt.ylabel("y-label")
```

```python
mat = np.arange(0, 100).reshape(10,10)
```

```python
plt.imshow(mat, cmap='coolwarm')
```

```python
mat = np.random.randint(0, 1000, (10, 10))
```

```python
plt.imshow(mat)
plt.colorbar()
```
